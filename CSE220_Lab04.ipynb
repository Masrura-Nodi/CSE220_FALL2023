```python
class Patient:
  def __init__(self, elem, next, pre):
    self.elem = elem
    self.next = next
    self.pre = pre
```

```python
class WRM:
    def __init__(self):
      self.n = Patient(None, None, None)
      self.n.next = self.n
      self.n.pre = self.n
      self.tail = self.n
    def RegisterPatient(self, id, name, age, blood_g):
        new_p_r = Patient(name, self.n, self.tail)
        self.tail.next = new_p_r
        self.tail = new_p_r
        self.n.pre = self.tail
    def ServePatient(self):
        if self.n.next == self.n:
            print('* NO PATIENT IN WAITING ROOM *')
        else:
          c_n = self.n.next
          p_n = c_n.pre
          n_n = c_n.next
          p_n.next = n_n
          n_n.pre = p_n
          c_n.next = None
          c_n.pre = None
    def ShowAllPatient(self):
        temp = self.n.next
        if temp == self.n :
          print(">>>> Waiting room is empty")
        else:
          while temp != self.n:
            print(' > ',temp.elem)
            temp = temp.next
          print("")
    def CanDoctorGoHome(self):
          if self.n.next == self.n:
            print(">>>> YES")
          else:
            print(">>>> NO")
    def CancelAll(self):
          self.n.next = self.n
          self.n.prev = self.n
          self.tail = self.n
          print(">>>> Appointments are cancelled")
    def ReverseTheLine(self):
          self.n.next, self.n.prev = self.n.prev, self.n.next
          temp = self.n.next
          while temp != self.n:
            temp.next,temp.pre = temp.pre,temp.next
            temp = temp.next
          print('>>>> The line is reversed ')
```

```python

wrm=WRM()
while(5):
    print("a Add Patient")
    print("b Serve Patient")
    print("c Show All patients")
    print("d Can Doctor go Home?")
    print("e Cancel all Appointment")
    print("f ReverseTheLine")
    print("g Exit")
    print("")
    print('Enter any option a > b > c > d > e > f > g: ')
    n = input()
    if n =='a':
        id = int(input("Enter patient ID : "))
        name = input("Enter patient Name : ")
        age = int(input("Enter patient Age : "))
        blood_g = input("Enter patient blood group : ")
        wrm.RegisterPatient(id, name, age, blood_g)
    elif n == 'b':
        print("# Serve a patient")
        wrm.ServePatient()
    elif n == 'c':
        print("# Show all patient")
        wrm.ShowAllPatient()
    elif n == 'd':
        print("# Can doctor go home?")
        wrm.CanDoctorGoHome()
    elif n == 'e':
        print("# Cancel all appointment")
        wrm.CancelAll()
    elif n == 'f':
        print("# Reverse the line")
        wrm.ReverseTheLine()
    elif n == 'g':
        print('EXIT')
        break
    else:
        print('ERROR')
```
